```python
from pandas.api.types import CategoricalDtype

cat_type= CategoricalDtype(categories=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'], ordered=True)

def create_features(df, label=None):
    """
    Creates time series features from datetime index
    """
    
    df= df.copy()
    df['date']=df.index
    df['hour']=df['date'].dt.hour
    df['dayofweek']=df['date'].dt.dayofweek
    df['weekday']=df['date'].dt.day_name()
    df['weekday']=df['weekday'].astype(cat_type)
    df['quarter']=df['date'].dt.quarter
    df['month']=df['date'].dt.month
    df['year']=df['date'].dt.year
    df['dayofyear']=df['date'].dt.dayofyear
    df['dayofmonth']=df['date'].dt.day
    df['weekofyear']=df['date'].dt.weekofyear
    df['date_offset'] = (df.date.dt.month*100+df.date.dt.day -320)%1300
    
    X=df[['hour','dayofweek','quarter','month','year','dayofyear','dayofmonth','weekofyear','weekday']]
    return X

X= create_features(data_new)
all_data= pd.concat([X,data_new],axis=1)
```

```python
# standard library
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import gc

# sklearn
from sklearn import model_selection, metrics, ensemble

# visualisation
import matplotlib.pyplot as plt
%matplotlib inline

# statsmodel
import statsmodels.api as sm
```

```python
data = pd.read_csv('delhi_aqi.csv', parse_dates=['date'])
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
data.reset_index(drop=True,inplace=True)
# feature extraction
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data['hour'] = data['date'].dt.hour

# set datetime as index
data = data.set_index('date')

# view
data.head()
```

```python
# standard library
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import gc

# sklearn
from sklearn import model_selection, metrics, ensemble

# visualisation
import matplotlib.pyplot as plt
%matplotlib inline

# statsmodel
import statsmodels.api as sm
```

```python
# custom function to get the regression of pollutants using StatModels API
def statmodel_reg(pollutant):
    # feature engineering & data-split
    x = data.iloc[:,-4:]
    y = data[[pollutant]]
    # time-series splitting of data
    tss = model_selection.TimeSeriesSplit()
    for train_index, test_index in tss.split(x):
        x_train, x_test = x[:len(train_index)], x[len(train_index): (len(train_index)+len(test_index))]
        y_train, y_test = y[:len(train_index)], y[len(train_index): (len(train_index)+len(test_index))]
    print(f"Training size = {len(x_train)} | Testing size = {len(x_test)}\n")
    # stats model building
    model = sm.OLS(np.asarray(y_train), np.asarray(x_train))
    results = model.fit()
    print(results.summary())
        # prediction
    y_pred = results.predict(np.asarray(x_test))
    mse = metrics.mean_squared_error(np.asarray(y_test), y_pred)
    rmse = mse**0.5
    print(f"\n\n *** Predicted: {rmse} | Average: {y.mean()[0]} ***")
    if (rmse < y.mean()[0]) or (rmse > y.mean()[0]):
        print("\nThe regressor model didn't perform so well...")
```

```python
# checking for pollutant = pm2_5
statmodel_reg('pm2_5')
```

# **Regression using Gradient Boosting**

One might also presume that the current 'pollutant' concentration is influenced by the amount that was previously in the air. In especially stagnant weather conditions, the 'pollutant' concentration may accumulate more than it would under more dynamic weather. A way to capture this influence is to introduce three new features:

1. The six-hour lagged 'pollutant' concentration
2. The twelve-hour lagged 'pollutant' concentration
3. The twenty-four-hour lagged 'pollutant' concentration
For any given hour, the six (twelve, twenty-four) hour lagged average is an average of the 'pollutant' concentration in the previous six (twelve, twenty-four) hours. This is a reasonable feature to implement in a real-life scenario, as I assume the previous 'pollutant' measurements are available immediately to the data pipeline. I’ve arbitrarily chosen three different time windows, but it may actually be sufficient to have just one.

```python
def gradientboosting(pollutant):
    temp = data.reset_index()
    six_df = pd.DataFrame()
    twelve_df = pd.DataFrame()
    tf_df = pd.DataFrame()
    
    for i in range(len(temp)):
        six_df = pd.concat([six_df,pd.DataFrame({'6hr_avg':[sum(temp[pollutant].iloc[i:i+6])/6]}, index = [i+6])])
        twelve_df = pd.concat([twelve_df,pd.DataFrame({'12hr_avg':[sum(temp[pollutant].iloc[i:i+12])/12]}, index = [i+12])])
        tf_df = pd.concat([tf_df,pd.DataFrame({'24hr_avg':[sum(temp[pollutant].iloc[i:i+24])/24]}, index = [i+24])])

    temp = pd.merge(temp, six_df, left_index=True, right_index=True)
    temp = pd.merge(temp, twelve_df, left_index=True, right_index=True)
    temp = pd.merge(temp, tf_df, left_index=True, right_index=True)
    
    temp.reset_index(drop=True,inplace=True)
    temp = temp.set_index('date')
    # feature engineering & data-split
    x = temp.iloc[:,-7:]
    y = temp[[pollutant]]

    # time-series splitting of data
    tss = model_selection.TimeSeriesSplit()

    for train_index, test_index in tss.split(x):
        x_train, x_test = x[:len(train_index)], x[len(train_index): (len(train_index)+len(test_index))]
        y_train, y_test = y[:len(train_index)], y[len(train_index): (len(train_index)+len(test_index))]

    print(f"Training size = {len(x_train)} | Testing size = {len(x_test)}")
    
    # Gradient Boosting Regressor
    xgb = ensemble.GradientBoostingRegressor(random_state=0)
    xgb.fit(x_train,y_train)
    xgb_score = xgb.score(x_test, y_test)
    print(f"\nThe return the coefficient of determination of the prediction is {xgb_score}")
    
    # prediction
    y_pred = xgb.predict(x_test)
    mse = metrics.mean_squared_error(np.asarray(y_test), y_pred)
    rmse = mse**0.5
    print(f"\n\n *** Predicted: {rmse} | Average: {y.mean()[0]} ***")
        # create new dataframe with actual & predicted values
    df_pred = x_test.copy()
    df_pred['actual'] = y_test
    df_pred['predicted'] = y_pred.tolist()
    
    # plot
    plt.figure(figsize=(15,10))
    plt.plot(df_pred['actual'], label='Actual Value')
    plt.plot(df_pred['predicted'], label='Predicted Value')
    plt.title("predicted v/s true value for "+ pollutant)
    plt.xlabel('date')
    plt.ylabel(pollutant)
    plt.legend()
    plt.show()
```

```python
# checking for pollutant = pm2_5
gradientboosting('pm2_5')
```
